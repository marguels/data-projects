# Palermo Street Map

```python
import geopandas as gpd
import osmnx as ox
import matplotlib.pyplot as plt
```

Using the osmnx library we fetch 

```python
tags = {'building': True}   
buildings = ox.geometries_from_place('Palermo, Italy', tags)
```

1. Set image properties.
    - You can choose _facecolor_ from https://xkcd.com/color/rgb/.
    - Check map colors on: https://matplotlib.org/stable/tutorials/colors/colormaps.html
2. Set coordinates to match Palermo.
   I found and adjusted box coordinates with [bboxfinder](http://bboxfinder.com/).

```python
fig, ax = plt.subplots(1, figsize=(15,15))
# fig.patch.set_facecolor('xkcd:taupe')
ax.set_aspect('equal')
ax.axis('off')

ax.set_xlim(13.310629, 13.414636)
ax.set_ylim(38.076476, 38.160304)

buildings.plot(cmap='viridis', ax=ax)
```
